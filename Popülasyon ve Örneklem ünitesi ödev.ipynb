# Popülasyon ve Örneklem ünitesi ödev1

```python
örneklem1 = []
örneklem_ortalamaları1 = []
for elemensayisi in range(1,50):
    eleman = np.random.normal(5,0.2,1000)[0]
    örneklem1.append(eleman)
    örneklem_ortalamaları1.append(np.mean(örneklem1))
```

```python
örneklem2 = []
örneklem_ortalamaları2 = []
for elemensayisi in range(1,1000):
    eleman = np.random.normal(4,0.3,1000)[0]
    örneklem2.append(eleman)
    örneklem_ortalamaları2.append(np.mean(örneklem2))
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(örneklem_ortalamaları1, label='Örneklem 1',color='yellow') 
plt.legend(loc='upper right')
plt.subplot(1,2,2)
plt.hist(örneklem_ortalamaları2, label='Örneklem 2',color='red') 
plt.legend(loc='upper right')
plt.show()
```

Görüldüğü üzere öreneklem2'de yaklaşık 930 civarı değer ortalamaya çok yakın çıktı yaklaşık %93 oranında,
örneklem1'de ise değer sayısı az olduğundan yaklaşık 25 değeri ortalamaya yakın çıktı yaklaşık %50 oranında 
sonuç olarak örneklemdeki değer sayısı arttıkça normal dağılıma çok yakın bir görüntü elde edilmektedir.

# Popülasyon ve Örneklem ünitesi ödev1

```python
#1 yazı, 2 tura
örneklem = []
örneklem_ortalamaları = []
for atış_sayısı in range(1,100):
    atış = np.random.randint(low=1,high=2, size=1)[0]
    örneklem.append(atış)
    örneklem_ortalamaları.append(np.mean(örneklem))
```

```python
def orneklem_plotla(örneklem_ortalaması,orneklem_sayisi, subplot_no):
    baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
    eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
    plt.subplot(subplot_no)
    plt.hlines(1.5, 0, orneklem_sayisi, colors= 'r' )
    plt.plot(örneklem_ortalaması[0:orneklem_sayisi])
    plt.ylim((0,3))
    plt.title("Örneklem Sayısı = {}".format(orneklem_sayisi), fontdict = baslik_font)
    plt.xlabel("Örneklem Sayısı", fontdict = eksen_font)
    plt.ylabel("Örneklem Ortalaması", fontdict = eksen_font)
    return plt
```

```python
plt.figure(figsize=(15,3), dpi=100)
orneklem_plotla(örneklem_ortalamaları,100, 131) 
# 131 --> 1 satır ve 3 sütunluk subplot, 1.subplot
orneklem_plotla(örneklem_ortalamaları, 500, 132)
orneklem_plotla(örneklem_ortalamaları, 2000, 133)
plt.show()
```

grafiklerdede görüldüpü üzere atış sayısı arttıkça yazı ve turanın gelme olasılığı eşitleniyor.
